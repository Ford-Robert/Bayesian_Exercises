## Linear Regression
$
\begin{align*}
y_i &\sim{} \mathcal N\left(\mathbf{x}_i^\top \boldsymbol \beta, \sigma^2\right) && i = 1,\cdots,n\\
 &\;\quad {} \textrm{Assumptions} \\
1. &\;\quad {} \mathbf{x}_i^\top \boldsymbol \beta && \textrm{Linear model form true} && \hat y_i = \mathbf{x}_i^\top \boldsymbol{\hat \beta} \textrm{ from least squares}&& \hat \beta = \underset{\beta}{\textrm{argmin}} \sum_{i=1}^n(y_i - \hat y_i)^2 = \underset{\beta}{\textrm{argmin}} \; \hat{\textbf{e}}^\top\hat{\textbf{e}} \\
2. &\;\quad {} \sigma^2 && \textrm{Homoskedastity holds} && \sigma^2\neq\sigma_{x_i}^2 \textrm{ a function of } \mathbf{x}_i\\
3. &\;\quad {} \mathbf{x}_i && \textrm{Perfectly measured $\mathbf{x}_i$} && \mathbf{x}_i \textrm{ has no "uncertainty" }&&(\textit{$\mathbf{x}_i$ has no distributional assumptions})\\
4. &\;\quad {} \epsilon_i=y_i-\mathbf{x}_i^\top \boldsymbol \beta && \textrm{Normality of error terms} && \textrm{and residuals } \hat e_i=y_i-\hat y_i \\
\end{align*}$


## Generative Modeling

$p(y_i | \mathbf{x}_i, \boldsymbol {\hat \beta}, \hat \sigma)$ estimated as $\mathcal N\left(\mathbf{x}_i^\top \boldsymbol {\hat \beta}, \hat \sigma^2\right)$ is a **predictive model** as opposed to a **generative model**

Integrating out the estimation uncertainty provides the **posterior predictive distribution**

$p(y_i | \mathbf{x}_i) = \int p(y_i | \mathbf{x}_i, \boldsymbol {\hat \beta}, \hat \sigma) \; p(\boldsymbol {\hat \beta}, \hat \sigma)\; d \boldsymbol {\hat \beta} \; d \hat \sigma $

**Generative modeling** means considering stochastic rather than fixed $\mathbf{x}_i$

$p(y_i, \mathbf{x}_i) = p(y_i | \mathbf{x}_i) p(\mathbf{x}_i)  = \int p(y_i | \mathbf{x}_i, \boldsymbol \theta) \; p(\mathbf{x}_i | \boldsymbol \theta) \; p(\boldsymbol \theta)\; d \boldsymbol {\theta}$

## Least Squares

$
\begin{align*}
\hat \beta = \underset{\beta}{\textrm{argmin}} \sum_{i=1}^n(y_i - \hat y_i)^2 &={}
\underset{\beta}{\textrm{argmin}} \; \hat{\textbf{e}}^\top\hat{\textbf{e}}\\
&={} \underset{\beta}{\textrm{argmin}} \; (\textbf{y}-\textbf{X}\boldsymbol \beta)^\top(\textbf{y}-\textbf{X}\boldsymbol \beta)\\
&={} \underset{\beta}{\textrm{argmin}} \; \textbf{y}^\top \textbf{y} -2 (\textbf{X}\boldsymbol \beta)^\top \textbf{y} + (\textbf{X}\boldsymbol \beta)^\top (\textbf{X}\boldsymbol \beta)\\
\\
0 &={} \nabla_\beta \left[ \textbf{y}^\top \textbf{y} -2 \boldsymbol \beta^\top \textbf{X}^\top \textbf{y} + \boldsymbol \beta^\top \textbf{X}^\top \textbf{X}\boldsymbol \beta \right]\\
 &={} -2  \textbf{X}^\top \textbf{y} + 2 \textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} \\
 &={} - \textbf{X}^\top \textbf{y} + \textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} \\
 \\
\textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} &={} \textbf{X}^\top \textbf{y}  \\
\boldsymbol {\hat \beta} &={} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y}  \\
\hat{\textbf{y}} = \textbf{X} \boldsymbol {\hat \beta} &={} \textbf{X} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y} = \textbf{H} \textbf{y} = \hat{\textbf{y}} \\
\\
\textbf{X}^\top (\textbf{y}-\hat{\textbf{y}}) &={} \textbf{X}^\top \textbf{y} - \textbf{X}^\top\textbf{X} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y} = \textbf{X}^\top \textbf{y} - \textbf{X}^\top \textbf{y} = \textbf{0} \quad  \Longrightarrow{} \quad \sum_{i=1}^n \hat e_i = 0 \\
&\Longrightarrow{} \quad \textbf{X}^\top \textbf{y} =   \textbf{X}^\top \hat{\textbf{y}}
\end{align*}$

## Generalized Linear Models (GLM)

1. Not all data has **normally distributed error terms**
2. $E[Y|\textbf{x}_i] = \textbf{x}_i^T \boldsymbol \beta$ can paramaterize more than a **normal distribution**

$E[Y|\textbf{x}_i] = e^{\textbf{x}_i^T \boldsymbol \beta }$ parameterizes $\lambda > 0$ of a **Poisson distribution**

> A **Poisson** GLM uses a **log link function** so $\log(\lambda_i) = \textbf{x}_i^T \boldsymbol \beta$

$E[Y|\textbf{x}_i] = \frac{1}{1 + e^{-\textbf{x}_i^T \boldsymbol \beta }}$ parameterizes $p \in [0,1]$ of a **Bernoulli distribution**

> A **logistic regression** uses a **logit (log odds) link function** so $\frac{p}{1-p_i} = \textbf{x}_i^T \boldsymbol \beta$

An alternative is the **probit link function** $\Phi^{-1}(p_i) = \textbf{x}_i^T \boldsymbol \beta$ where $\Phi^{-1}$ is the inverse CDF of a standard normal distribution; thus, $E[y|\textbf{x}_i] = p_i = \Phi(\textbf{x}_i^T \boldsymbol \beta) = Pr(z_i \geq 0) \; \textrm{ for } \; z_i \sim \mathcal N(\textbf{x}_i^T \boldsymbol \beta,1)$

$
\begin{align*}
\Phi(\textbf{x}_i^T \boldsymbol \beta)\sqrt{2\pi} &={} \int_{-\infty}^{x_i^\top\boldsymbol{\beta}_{p\times 1}} e^{-\frac{1}{2}z^2_i}dz\\
&={} \int_{-\infty}^{0} e^{-\frac{1}{2}(z_i + x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz = \int_{-\infty}^{0} e^{-\frac{1}{2}(-z_i - x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz =  \int^{\infty}_{0} e^{-\frac{1}{2}(z_i-x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz
\end{align*}$

This formulation maps **binary** $y_i$ to a **latent normally distributed representation** $z_i$. This demonstrates a fundamental idea underlying **copulas** which transform arbitrary distribution into latent normal distributions (several of which) which may then be tied together as a multivariate normal distribution to provide inference on an underlying **covariance structure**.

- The general **copula** formulation for $j^{th}$ random variable $\textbf{Y}_j$ with CDF $F_{Y_j}$ is $\Phi^{-1}(F_{Y_j}(y_{ij}))$ but the trick used above is to instead map this as $\Phi^{-1}(F_{Y_j}(E[Y_{ij}|\textbf{x}_i|))$

## Shrinkage Regularization

**Machine Learning** fits models by optimizing penalized **loss functions**.

Two classic regularizations are "ridge" and "lasso" regression, which respectively use $L_2$ and $L_1$ penalty functions.

- Ridge: $$\sum_{i=1}^n \frac{1}{2}(y_i-x_i^T\beta_{p \times 1})^2 + \lambda \sum_{j=1}^n \beta_j^2 = \frac{1}{2}(y-X\beta)^T(y-X\beta) + \lambda \sum_{j=1}^n \beta_j^2 = \frac{1}{2}||y-X\beta||_2^2 + \lambda ||\beta||_2^2 $$
- Lasso: $$\sum_{i=1}^n \frac{1}{2}(y-x_i^T\beta_{p \times 1})^2 + \lambda \sum_{j=1}^n |\beta_j| = \frac{1}{2}(y-X\beta)^T(y-X\beta) + \lambda \sum_{j=1}^n |\beta_j| = \frac{1}{2}||y-X\beta||_2^2 + \lambda ||\beta||_1$$

### As priors

Bayes has been regularizing since [1763](https://en.wikipedia.org/wiki/Thomas_Bayes#Bayes'_theorem)

- **Machine Learning** is the original inventor or "regularization"?
- Bayesian analysis has been regularizing since it's inception: *priors are regularization*


**Ridge Regression** **$L_2$-penalty** is just a MAP estimate using **norml priors** for the **regression coefficients**

\begin{align*}
 \beta_i \sim {} & \text{Normal}(b_i,s_i) & f(\beta_i| b_i, s_i) = {}& \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{\beta_i-b_i}{s_i}\right)^2} & \overbrace{\underbrace{|\beta_i-b_i|^2}}_{\text{Squared Penalization}}^{L_2} \\
y_i \sim {} & \text{Normal}(x_i^T\beta, \sigma) & \sigma \sim {} & \text{HalfNormal}(\sigma_0)
\end{align*}

**Lasso Regression** **$L_1$-penalty** is just replacing the **normal priors** with **Laplace ("Double Exponential") prior distributions**

\begin{align*}
\beta_i \sim {} & \text{Laplace}(b_i,s_i) & f(\beta_i| b_i, s_i) = {}& {\frac  {1}{2b}}\exp \left(-{\frac{|\beta_i-b_i |}{s_i}}\right) & \overbrace{\underbrace{|\beta_i-b_i|}}_{\text{Absolute Penalization}}^{L_1}\\
y_i \sim {} & \text{Normal}(x_i^T\beta, \sigma) & \sigma \sim {} & \text{HalfNormal}(\sigma_0)
\end{align*}

## Standard Linear Regression Model

We begin with the standard linear regression model

$$
y = X\beta + \epsilon, \quad \epsilon \sim N(0, \sigma^2 I),
$$

so that the likelihood is

$$
p(y \mid X, \beta, \sigma^2) \propto \exp\!\left(-\frac{1}{2\sigma^2}(y - X\beta)^\top (y - X\beta)\right).
$$

---

## 1. Posterior for $\beta$ Given $\sigma^2$

### Prior:

Assume

$$
\beta \sim N(\beta_0, \Sigma_\beta).
$$

### Likelihood:

We have

$$
p(y \mid X, \beta, \sigma^2) \propto \exp\!\left(-\frac{1}{2\sigma^2}(y - X\beta)^\top (y - X\beta)\right).
$$

### Posterior:

By Bayes’ rule (ignoring normalizing constants), the posterior is

$$
p(\beta \mid y, X, \sigma^2) \propto p(y \mid X, \beta, \sigma^2)\, p(\beta).
$$

Substitute the likelihood and the prior:

$$
p(\beta \mid y, X, \sigma^2) \propto \exp\!\left(-\frac{1}{2\sigma^2}(y - X\beta)^\top (y - X\beta)\right)
\exp\!\left(-\frac{1}{2}(\beta - \beta_0)^\top \Sigma_\beta^{-1} (\beta - \beta_0)\right).
$$

### Combine the Exponents:

Write the exponent as

$$
-\frac{1}{2}\left[\frac{1}{\sigma^2}(y - X\beta)^\top (y - X\beta) + (\beta - \beta_0)^\top \Sigma_\beta^{-1} (\beta - \beta_0)\right].
$$

Expanding the quadratic term from the likelihood:

$$
(y - X\beta)^\top (y - X\beta) = y^\top y - 2\beta^\top X^\top y + \beta^\top X^\top X \beta.
$$

The prior term expands as

$$
(\beta - \beta_0)^\top \Sigma_\beta^{-1} (\beta - \beta_0)
= \beta^\top \Sigma_\beta^{-1} \beta - 2\beta_0^\top \Sigma_\beta^{-1}\beta + \beta_0^\top \Sigma_\beta^{-1}\beta_0.
$$

Ignoring terms not involving $\beta$, the exponent that involves $\beta$ becomes

$$
-\frac{1}{2}\left[\beta^\top\left(\frac{1}{\sigma^2}X^\top X + \Sigma_\beta^{-1}\right)\beta - 2\beta^\top\left(\frac{1}{\sigma^2}X^\top y + \Sigma_\beta^{-1}\beta_0\right)\right].
$$

### Completing the Square:

This quadratic form can be written in the standard form for a multivariate normal density:

$$
-\frac{1}{2}(\beta - \beta_N)^\top \Sigma_N^{-1} (\beta - \beta_N),
$$

where

$$
\Sigma_N = \left(\frac{1}{\sigma^2}X^\top X + \Sigma_\beta^{-1}\right)^{-1},
$$

and

$$
\beta_N = \Sigma_N\left(\frac{1}{\sigma^2}X^\top y + \Sigma_\beta^{-1}\beta_0\right).
$$

Thus, the posterior for $\beta$ is

$$
\beta \mid y, X, \sigma^2 \sim \mathcal{N}(\beta_N, \Sigma_N).
$$

---

## 2. Posterior for $\sigma^2$ with an Inverse-Gamma Prior

Now, suppose we have an inverse-gamma prior for $\sigma^2$ given by

$$
\sigma^2 \sim \text{IG}(\alpha^*, \beta^*),
$$

with density (ignoring constants that do not depend on $\sigma^2$)

$$
p(\sigma^2) \propto (\sigma^2)^{-(\alpha^* + 1)} \exp\!\left(-\frac{\beta^*}{\sigma^2}\right).
$$

For simplicity, consider the likelihood for $y$ given $\beta$ and $\sigma^2$:

$$
p(y \mid X, \beta, \sigma^2) \propto (\sigma^2)^{-n/2} \exp\!\left(-\frac{1}{2\sigma^2}(y - X\beta)^\top (y - X\beta)\right),
$$

where $n$ is the number of observations.

### Posterior (Conditional on $\beta$):

Combining the likelihood with the prior, we have

$$
p(\sigma^2 \mid y, X, \beta) \propto p(y \mid X, \beta, \sigma^2) \, p(\sigma^2).
$$

Substitute the expressions:

$$
p(\sigma^2 \mid y, X, \beta) \propto (\sigma^2)^{-n/2}\exp\!\left(-\frac{1}{2\sigma^2}S\right)
\; (\sigma^2)^{-(\alpha^* + 1)}\exp\!\left(-\frac{\beta^*}{\sigma^2}\right),
$$

where we have defined

$$
S = (y - X\beta)^\top (y - X\beta).
$$

### Combine the Powers and Exponents:

$$
p(\sigma^2 \mid y, X, \beta) \propto (\sigma^2)^{-\left(\alpha^* + \frac{n}{2} + 1\right)} \exp\!\left(-\frac{1}{\sigma^2}\left[\beta^* + \frac{1}{2} S\right]\right).
$$

This is recognized as the kernel of an inverse-gamma distribution. That is, the posterior for $\sigma^2$ is

$$
\sigma^2 \mid y, X, \beta \sim \text{IG}\!\left(\alpha^* + \frac{n}{2}, \; \beta^* + \frac{1}{2}(y - X\beta)^\top (y - X\beta)\right).
$$


```python
import kagglehub
import pandas as pd
import os
import pymc as pm
import arviz as az
import numpy as np

#Q2

# Download latest version
path = kagglehub.dataset_download("mahatiratusher/heart-disease-risk-prediction-dataset")

file_name = 'heart_disease_risk_dataset_earlymed.csv'
data_file = os.path.join(path, file_name)

# Load the CSV file into a DataFrame
df = pd.read_csv(data_file)

# Print the first few rows of the DataFrame
# print(df.head())


X = df[['Smoking', 'High_BP', 'Chest_Pain', 'Age']].values
y = df['Heart_Risk'].values
p = X.shape[1]

with pm.Model() as model:
    sigma2 = pm.InverseGamma('sigma2', alpha=2, beta=1)
    sigma = pm.Deterministic('sigma', sigma2**0.5)

    betas = pm.Normal('betas', mu=0, sigma=1, shape=p)

    mu = pm.math.dot(X, betas)
    y_obs = pm.Normal('y', mu=mu, sigma=sigma, observed=y)

    trace = pm.sample(2000, tune=1000, return_inferencedata=True)
```

```python
#Q2

import matplotlib.pyplot as plt

summary = az.summary(trace, hdi_prob=0.95)
print(summary)

# Plot trace plots to assess convergence and mixing of the chains
az.plot_trace(trace)
plt.show()

# Plot posterior distributions with 95% credible intervals highlighted
az.plot_posterior(trace, hdi_prob=0.95)
plt.show()
```

```python
#Q3
df = df.iloc[:1000]

np.random.seed(42)
n_obs = df.shape[0]
outlier_indices = np.random.choice(n_obs, size=int(0.05 * n_obs), replace=False)
df.loc[outlier_indices, 'Heart_Risk'] += np.random.normal(10, 2, size=len(outlier_indices))

# Define design matrix X and response y
# (Update column names if necessary.)
X = df[['Smoking', 'High_BP', 'Chest_Pain', 'Age']].values
y = df['Heart_Risk'].values
n, p = X.shape


with pm.Model() as robust_model:
    betas = pm.Normal('betas', mu=0, sigma=1, shape=p)

    tau = pm.Gamma('tau', alpha=1, beta=1)
    v = pm.Exponential('v', lam=1/10) + 2  # ensures v > 2

    tau_i = pm.Gamma('tau_i', alpha=v/2, beta=v/2, shape=n)

    mu = pm.math.dot(X, betas)

    sigma_i = 1 / pm.math.sqrt(tau * tau_i)

    y_obs = pm.Normal('y', mu=mu, sigma=sigma_i, observed=y)

    trace_robust = pm.sample(2000, tune=1000, return_inferencedata=True, target_accept=0.95)
```

```python
tau_i_summary = az.summary(trace_robust, var_names=["tau_i"], hdi_prob=0.95)
print("Summary for tau_i (latent weights):")
print(tau_i_summary)

# Compute the median of tau_i across chains and draws for each observation
tau_i_medians = np.median(trace_robust.posterior["tau_i"].values, axis=(0, 1))
# Compute the 25th percentile of these medians
q25 = np.percentile(tau_i_medians, 25)
# Flag observations with median tau_i below the 25th percentile as potential outliers
outlier_flags = tau_i_medians < q25

print("\nObservations flagged as potential outliers (lower 25% of tau_i):")
for i, flag in enumerate(outlier_flags):
    if flag:
        print(f"Observation {i}: median tau_i = {tau_i_medians[i]:.3f}")

# -----------------------------------------------------
# 6. Compute Posterior Influence via the Hat Matrix
# -----------------------------------------------------
# Stack the posterior samples for tau and tau_i to access the sample dimension correctly
posterior_tau = trace_robust.posterior["tau"].stack(sample=("chain", "draw"))
posterior_tau_i = trace_robust.posterior["tau_i"].stack(sample=("chain", "draw"))
total_samples = posterior_tau.sizes["sample"]

n_samples = 100  # Choose 100 posterior samples for illustration
sample_idxs = np.random.choice(total_samples, n_samples, replace=False)

# Preallocate an array to store the hat matrix diagonal for each sample
hat_diag_samples = np.zeros((n_samples, n))

# Loop over selected posterior samples and compute the hat matrix diagonal elements
for j, idx in enumerate(sample_idxs):
    # Extract the sample for tau (a scalar) and tau_i (a vector of length n)
    tau_sample = posterior_tau.values[idx]
    tau_i_sample = posterior_tau_i.values[:, idx]  # Correctly indexing the sample dimension

    # Compute observation-specific weights: w_i = tau * tau_i
    w = tau_sample * tau_i_sample
    W = np.diag(w)

    # Compute the weighted hat matrix:
    # H = sqrt(W) @ X @ inv(X.T @ W @ X) @ (sqrt(W) @ X).T
    sqrtW = np.diag(np.sqrt(w))
    XWX = X.T @ W @ X
    inv_XWX = np.linalg.inv(XWX)
    H = sqrtW @ X @ inv_XWX @ (sqrtW @ X).T
    hat_diag_samples[j, :] = np.diag(H)

# -----------------------------------------------------
# 7. Compare Influence of an Outlier vs. a Non-Outlier
# -----------------------------------------------------
# Select one observation flagged as an outlier and one not flagged
outlier_idx = np.where(outlier_flags)[0][0]
non_outlier_idx = np.where(~outlier_flags)[0][0]

plt.figure(figsize=(8, 4))
plt.hist(hat_diag_samples[:, outlier_idx], bins=30, alpha=0.6, label=f"Obs {outlier_idx} (Outlier)")
plt.hist(hat_diag_samples[:, non_outlier_idx], bins=30, alpha=0.6, label=f"Obs {non_outlier_idx} (Non-Outlier)")
plt.xlabel("Influence (hat diag)")
plt.ylabel("Frequency")
plt.legend()
plt.title("Posterior Distributions of Influence for Selected Observations")
plt.show()


summary = az.summary(trace_robust, hdi_prob=0.95)
print("Posterior Summary with 95% Credible Intervals:")
print(summary)

# Plot trace plots to assess the sampling convergence and mixing of the chains
az.plot_trace(trace_robust)
plt.show()

# Plot posterior distributions for all parameters with 95% credible intervals highlighted
az.plot_posterior(trace_robust, hdi_prob=0.95)
plt.show()
```
